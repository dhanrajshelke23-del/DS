```python
#1.Extract Sample document and apply following document preprocessing methods: Tokenization, POS
#Tagging, stop words removal, Stemming and Lemmatization.
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('omw-1.4')
```

```python
text="Hello. What is your name?"
```

```python
from nltk.tokenize import sent_tokenize
tokenized_text=sent_tokenize(text)
```

```python
tokenized_text
```

```python
from nltk.tokenize import word_tokenize
tokenized_word=word_tokenize(text)
tokenized_word
```

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
stop_words
```

```python
wordsFiltered = []
for w in tokenized_word:
    if w not in stop_words:
        wordsFiltered.append(w)
print(wordsFiltered)
```

```python
from nltk.stem import PorterStemmer
words = ["play" , "plays" , "playing" , "played"]
ps = PorterStemmer()
for word in words:
    print(ps.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
text1 = "cars children studies runs"
tokenization = nltk.word_tokenize(text1)
for w in tokenization:
    print("Lemma for {} is {}".format(w,wordnet_lemmatizer.lemmatize(w)))
```

```python
for word in tokenized_word:
    print(nltk.pos_tag([word]))
```

```python
#2. Create representation of documents by calculating Term Frequency and Inverse Document Frequency.
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documentA = 'Python is a powerful programming language'
documentB = 'Java is widely used in software development'
```

```python
bagOfWordsA = documentA.split(' ')
bagOfWordsB = documentB.split(' ')
```

```python
bagOfWordsA
```

```python
bagOfWordsB
```

```python
uniqueWords = set(bagOfWordsA).union(set(bagOfWordsB))
```

```python
uniqueWords
```

```python
numOfWordsA = dict.fromkeys(uniqueWords, 0)
for word in bagOfWordsA:
    numOfWordsA[word] += 1
numOfWordsB = dict.fromkeys(uniqueWords, 0)
for word in bagOfWordsB:
    numOfWordsB[word] += 1
```

```python
numOfWordsA
```

```python
numOfWordsB
```

```python
#Term Frequency (TF)
```

```python
def computeTF(d, bag):
    return {w: c/len(bag) for w, c in d.items()}

tfA = computeTF(numOfWordsA, bagOfWordsA)
tfB = computeTF(numOfWordsB, bagOfWordsB)
```

```python
tfA
```

```python
tfB
```

```python
#Inverse Document Frequency (IDF)
```

```python
import math
def computeIDF(docs):
    N = len(docs)
    idf = dict.fromkeys(docs[0], 0)
    for doc in docs:
        for w, v in doc.items():
            if v > 0: idf[w] += 1
    return {w: math.log(N/v) for w, v in idf.items()}

idfs = computeIDF([numOfWordsA, numOfWordsB])
```

```python
idfs
```

```python
def computeTFIDF(tf, idf):
    return {w: tf[w]*idf[w] for w in tf}

tfidfA = computeTFIDF(tfA, idfs)
tfidfB = computeTFIDF(tfB, idfs)
```

```python
tfidfA
```

```python
tfidfB
```

```python
tfidfA = computeTFIDF(tfA, idfs)
tfidfB = computeTFIDF(tfB, idfs)
df = pd.DataFrame([tfidfA, tfidfB])
df
```
