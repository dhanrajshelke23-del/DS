```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.info()
df.describe()
```

```python
df.dtypes
```

```python
# Histogram for Sepal Length

plt.figure(figsize=(7,5))

sns.histplot(df['sepal_length'],
             bins=20,
             kde=True,
             color='skyblue')

plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")

plt.grid()
plt.show()
```

```python
# Histogram for Sepal Width

plt.figure(figsize=(7,5))

sns.histplot(df['sepal_width'],
             bins=20,
             kde=True,
             color='orange')

plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Width")

plt.grid()
plt.show()
```

```python
# Histogram for Petal Length

plt.figure(figsize=(7,5))

sns.histplot(df['petal_length'],
             bins=20,
             kde=True,
             color='green')

plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Length")

plt.grid()
plt.show()
```

```python
# Histogram for Petal Width

plt.figure(figsize=(7,5))

sns.histplot(df['petal_width'],
             bins=20,
             kde=True,
             color='purple')

plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Width")

plt.grid()
plt.show()
```

```python
# Boxplot for Sepal Length
plt.figure(figsize=(7,5))
sns.boxplot(y='sepal_length', data=df, color='skyblue')

plt.ylabel("Sepal Length")
plt.title("Boxplot of Sepal Length")

plt.grid()
plt.show()
```

```python
# Boxplot for Sepal Width
plt.figure(figsize=(7,5))
sns.boxplot(y='sepal_width', data=df, color='orange')

plt.ylabel("Sepal Width")
plt.title("Boxplot of Sepal Width")

plt.grid()
plt.show()
```

```python
# Boxplot for Petal Length
plt.figure(figsize=(7,5))
sns.boxplot(y='petal_length', data=df, color='green')

plt.ylabel("Petal Length")
plt.title("Boxplot of Petal Length")

plt.grid()
plt.show()
```
